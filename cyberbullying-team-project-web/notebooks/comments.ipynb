The data generated using ChatGPT was saved into a CSV file with the following columns:
comment_text – the content of the comment
comment_status – the initial classification label (e.g., positive, neutral, or negative)
comment_id – a unique identifier for each comment
This CSV file was then imported into R for further analysis.

```r
data <- read.csv("../data/comments3.csv", fileEncoding = "UTF-8")
head(data, 10)
```

Using the randomNames package, fake names were generated for each comment. The names were formatted as FirstNameLastName and added to a new column called comment_fake_user. The final dataset was then exported as a CSV file with UTF-8 encoding and BOM to ensure correct display in Excel.

```r
# Install if not already installed
if (!require(randomNames)) {
  install.packages("randomNames")
  library(randomNames)
}

fake_names <- randomNames(nrow(data)) 


cleaned_names <- sapply(fake_names, function(name) {
  parts <- strsplit(name, ",")[[1]]
  paste0(parts[2], parts[1])  
})

data$comment_fake_user <- cleaned_names

head(data,10)
```

```r
write_utf8_bom <- function(data, file_path) {

  con <- file(file_path, open = "wb")
  

  writeBin(charToRaw("\xEF\xBB\xBF"), con)
  

  write.csv(data, con, row.names = FALSE)
  

  close(con)
}
```
